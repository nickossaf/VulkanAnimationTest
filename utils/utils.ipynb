# create vertex list

```python
x = np.linspace(0, 1, 33)
y = np.linspace(0, 1, 33)
x_coord = (x - 0.5) / 2
y_coord = (y - 0.5) / 2
xx_coord, yy_coord = np.meshgrid(x_coord, y_coord)
xx, yy = np.meshgrid(x, y)
color_r = xx
color_g = np.zeros_like(xx)
color_b = yy
res = np.stack([xx_coord.flatten(), yy_coord.flatten(), \
                #color_r.flatten(), color_g.flatten(), color_b.flatten(), \
                xx.flatten(), yy.flatten()], axis=0).T
res.shape
```

```python
with open('../resource/vertex2.txt', 'w') as f:
    for row in res:
        for i in row:
            f.write(str(i) + ' ')
        f.write('\n')
```

# create vertex index list

```python
def lin(x, y):
    return y * 33 + x
```

```python
idx = []
for i in range(33):
    for j in range(33):
        if i > 0 and j > 0:
            idx.append([lin(i, j), lin(i-1, j), lin(i, j-1)])
        if i < 32 and j < 32:
            idx.append([lin(i, j), lin(i+1, j), lin(i, j+1)])
idx
```

```python
with open('../resource/index.txt', 'w') as f:
    for row in idx:
        for i in row:
            f.write(str(i) + ' ')
        f.write('\n')
```
